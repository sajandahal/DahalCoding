```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have both tesla_stock and tesla_revenue dataframes
# Replace the data with your actual dataframes
tesla_stock_data = {
    'Date': ['2010-06-29', '2010-06-30', '2010-07-01', '2010-07-02', '2010-07-06'],
    'Close': [1.592667, 1.588667, 1.464, 1.28, 1.074],
}

tesla_revenue_data = {
    'Date': ['2010-06-29', '2010-06-30', '2010-07-01', '2010-07-02', '2010-07-06'],
    'Revenue': [1000000, 1200000, 900000, 800000, 950000],
}

tesla_stock = pd.DataFrame(tesla_stock_data)
tesla_revenue = pd.DataFrame(tesla_revenue_data)

# Define the make_dual_graphs function
def make_dual_graphs(stock_df, revenue_df, title):
    plt.figure(figsize=(12, 8))

    # Plot the stock prices
    plt.subplot(2, 1, 1)
    plt.plot(stock_df['Date'], stock_df['Close'], label='Close Price', marker='o', color='blue')
    plt.title(f'{title} - Historical Share Price')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.grid(True)

    # Plot the revenue
    plt.subplot(2, 1, 2)
    plt.plot(revenue_df['Date'], revenue_df['Revenue'], label='Revenue', marker='o', color='orange')
    plt.title(f'{title} - Historical Revenue')
    plt.xlabel('Date')
    plt.ylabel('Revenue')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Call the make_dual_graphs function with your data and a title
make_dual_graphs(tesla_stock, tesla_revenue, title='Tesla Historical Data')
```
